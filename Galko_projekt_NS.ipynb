```python
#import knižníc pandas, numpy a pyplot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#načítanie vstupného datasetu
dataset=pd.read_csv('AguaH.csv')
dataset
```

```python
#vypísanie oblastí s ktorými budem pracovať
dataset["TU"].unique()
```

```python
#vymazanie riadkov, ktoré obsahujú prázdnu hodnotu (pre lahšiu prácu s df)
dataset = dataset.dropna(how='any',axis=0) 
```

```python
#výber sĺpcov podľa názvu s ktorými chcem pracovať
pom=(dataset[['f.1_ENE_09','f.1_JUN_09','f.1_ENE_15','f.1_JUN_15','TU']])
pom
```

```python
#výber 1000 údajov z danej oblasti na základe porovnania a následne uloženie do jednotlivých dataframe-ov
DataFrame0=(pom[pom["TU"]=="DOMESTICO MEDIO"][:1000])
DataFrame1=(pom[pom["TU"]=="DOMESTICO RESIDENCIAL"][:1000])
DataFrame2=(pom[pom["TU"]=="DOMESTICO BAJA"][:1000])
```

```python
#spojenie jednotlivých dataframe-ov do výsledného dataframu s ktorým budem pracovať
df_final=pd.concat((DataFrame0,DataFrame1,DataFrame2))
```

```python
#premenovanie stĺpcov do zrozumiteľnej podoby
df_final =(df_final.rename(columns={"f.1_ENE_09": "Feb_09", "f.1_ENE_15": "Feb_15", "f.1_JUN_09": "Jun_09", "f.1_JUN_15": "Jun_15", "TU":"Nazov"}))
df_final
```

```python
# načítanie triedy Perceptron
class Perceptron():   
    def __init__(self, rychlost_ucenia=0.01, pocet_iteracii=15):
        self.eta = rychlost_ucenia 
        self.n_iter = pocet_iteracii      
    
    def fit(self, X,y):
        self.vahy = np.zeros(1+ X.shape[1])
        self.errors_=[]
        
        self.vahy_to_visualize_1=[]
        self.vahy_to_visualize_2=[]
        self.vahy_to_visualize_3=[]
        
        for i in range(self.n_iter):
            errors = 0
            for matica,vystup in zip(X,y):
                update = self.eta * (vystup-self.predikcia(matica))
                self.vahy[1:] += update*matica
                self.vahy[0]+= update
                
                print(self.vahy[1:])
                
                errors += int(update !=0.0)
            self.errors_.append(errors)   
            self.vahy_to_visualize_1.append(self.vahy[0])
            self.vahy_to_visualize_2.append(self.vahy[1])
            self.vahy_to_visualize_3.append(self.vahy[2])
        return self  

    def net_input(self, X):
        return np.dot(X, self.vahy[1:])+ self.vahy[0]

    def predikcia(self, X):
        return np.where(self.net_input(X) >= 0.0, 1, -1)
```

```python
# porovnanie mesiaca Február v roku 2009 a 2015 v domoch strednej veľkosti
X = df_final.iloc[0:3000, [0,2]].values
y = df_final.iloc[0:3000, 4].values

print(X)
print(y)

y = np.where(y == 'DOMESTICO MEDIO', -1, 1)
print(y)
```

```python
moj_perceptron_domestico_medio_feb = Perceptron()
```

```python
moj_perceptron_domestico_medio_feb.fit(X,y)
```

```python
# vykreslenie chyby v domoch strednej veľkosti
plt.plot(moj_perceptron_domestico_medio_feb.errors_, color='red', label='DOMESTICO_MEDIO_FEB')

plt.legend()
```
